```python
import pandas as pd

# Define input and output file names
parquet_file = 'your_file.parquet'
csv_file = 'output.csv'

# Read the Parquet file into a pandas DataFrame
# The engine="pyarrow" parameter ensures pyarrow is used
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet', engine="pyarrow")

# Convert the DataFrame to a CSV file
# index=False prevents pandas from writing the DataFrame index to the CSV file
df.to_csv(csv_file, index=False)

print(f"Successfully converted {parquet_file} to {csv_file}")

# csv can be found at 
# "https://raw.githubusercontent.com/Pperez4/DATA_ENG_ZOOMCAMP_2026/main/output.csv"
```
